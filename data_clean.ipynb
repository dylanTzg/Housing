```python
import pandas as pd
```

```python
# Charger les données
data = pd.read_csv("data_clean.csv")

# Afficher les types de données uniques pour chaque colonne
for column in data.columns:
    unique_values = data[column].unique()
    print(f"Colonne: {column}")
    print(f"Types uniques: {unique_values[:10]}")  # Limiter l'affichage à 10 valeurs uniques pour plus de lisibilité
    print(f"Nombre de types uniques: {len(unique_values)}\n")
```

```python
# Filtrer les lignes ayant type_vente = 'Vente'
data = data[data['type_vente'] == 'Vente']

# Supprimer les colonnes ville, nom_rue, num_rue
data = data.drop(columns=['ville', 'nom_rue', 'num_rue'])
```
